```python
df_last.isnull().sum()
df_last.isna().sum()
```

```python
# 데이터 타입 변경
# 분양가격(㎡)의 타입이 Object, 문자열 타입이라 계산입 불가능. 따라서 계산이 가능한 수치 데이터로 변경시켜줘야한다.

# df_apt_price["분양가격(㎡)"]
# df_apt_price["분양가격(㎡)"].mean()
# df_apt_price["분양가격(㎡)"].astype(int)
df_last["분양가격"] = pd.to_numeric(df_last["분양가격(㎡)"], errors='coerce')
# type(pd.np.nan)
df_last["분양가격"].mean()
```

```python
# 각 데이터간의 단위 싱크 맞춰주기
# 전국 평균 분양 가격은 전용면적별 제곱미터당 평균분양가격

df_last["평당분양가격"] = df_last["분양가격"] * 3.3
df_last
```

```python
df_last["분양가격(㎡)"].describe()

# 문자형(Object) 데이터 요약
# count: 전체 데이터 갯수 (공백 문자 데이터도 갯수에 포함)
# unique: 중복되지 않은 값의 갯수
# top: 가장 빈번하게 등장한 문자
# frqe: 가장 빈번하게 등장한 문자의 갯수
```

```python
df_last["분양가격"].describe()

# 수치형 데이터 요약
# count: 전체 데이터 갯수
# mean: 평균
# std: 표준편차
# min, max: 최소값, 최대값
# 25%, 50%, 75%: 1,2,3사분위

# 평균값과 중앙값의 차이가 큰 이유는 최대값이 커서 그런 것 같다.
```

```python
# 규모구분을 전용면적 컬럼으로 변경
df_last['규모구분'].unique()
```

```python
# df_last['규모구분'].replace("전용면적", "") # 파이썬의 replace는 문자열이 완전히 일치해야 replace 함수가 동작.

# (1) 전용면적 문자 제거
df_last['전용면적'] = df_last['규모구분'].str.replace("전용면적", "")

# (2) 초과-> ~ / 이하 -> 제거
df_last['전용면적'] = df_last['전용면적'].str.replace("초과", "~")
df_last['전용면적'] = df_last['전용면적'].str.replace("이하", "")
# df_last['전용면적'] = df_last['전용면적'].str.replace(" ", "") # 빈 문자열 제거
df_last['전용면적'] = df_last['전용면적'].str.replace(" ", "").str.strip() # 빈 문자열 제거 + 앞뒤 문자열 공백 제거
```

```python
# 사용하지 않는 컬럼 제거하기
# axis 0: 행, 1: 열

df_last = df_last.drop(["규모구분", "분양가격(㎡)"], axis=1)
df_last
```

### pandas의 melt를 이용해서 데이터 형태 바꾸기

```python
df_first
```

```python
df_first.melt()
df_first_melt = df_first.melt(id_vars="지역", var_name="기간", value_name="평당분양가격")
```

```python
df_first_melt.columns = ["지역명", "기간", "평당분양가격"]
df_first_melt.head(1)
```

### 기간의 연도와 월 분리하기

```python
date = "2013년12월"
date.split("년")
# date.split("년")[0]
# date.split("년")[1].replace("월", "")
```

```python
def parse_year(date):
    year = date.split("년")[0]
    return int(year)

def parse_month(date):
    month = date.split("년")[1].replace("월", "")
    return int(month)
```

```python
df_first_melt["연도"] = df_first_melt["기간"].apply(parse_year)
df_first_melt["월"] = df_first_melt["기간"].apply(parse_month)

df_first_melt
```

```python
df_last
```

```python
# df_first에는 전용면적 데이터가 없기 때문에 df_last의 전용면적이 전체인 데이터만 가져와서 비교해봅시다.
cols = ["지역명", "연도", "월", "평당분양가격"]
df_last_prepare = df_last.loc[df_last["전용면적"]=="전체", cols]

df_last_prepare.head(1)
```

```python
df_first_prepare = df_first_melt[cols]
df_first_prepare
```

### concat으로 2개의 df 합치기

```python
df = pd.concat([df_first_prepare, df_last_prepare])
df
```
