```python
# 필요없는 행 제외 후 데이터 불러오기
seoul = pd.read_excel("3.seoul_popular_data.xls", header=2, usecols="B, D, G, J, N")
seoul
```

```python
# 컬럼 이름 변경
seoul.columns = ['구', '인구수', '한국인', '외국인', '고령자']
seoul
```

```python
seoul.drop([0], inplace=True)
seoul
```

```python
# 각 구별 전체 인구를 이용해서 구별 '외국인비율'과 '고령자비율' 컬럼 추가
seoul['외국인비율'] = round(seoul['외국인'] / seoul['인구수'] * 100, 1) #  소수점 첫째자리 반올림
seoul['고령자비율'] = round(seoul['고령자'] / seoul['인구수'] * 100, 1) #  소수점 첫째자리 반올림
seoul
```

```python
# 인구수로 정렬(내림차순) 
seoul.sort_values(by='인구수', ascending=True)
```

```python
seoul.sort_values(by='외국인', ascending=True)
```

```python
seoul.sort_values(by='외국인비율', ascending=True)
```

```python
seoul.sort_values(by='고령자', ascending=True)
```

```python
seoul.sort_values(by='고령자비율', ascending=True)
```

```python
# 인덱스 값을 구별 데이터로 변경
seoul.set_index('구', inplace=True)
seoul
```

```python
seoul['인구수'].plot(kind='bar', grid=True)
```

```python
seoul['인구수'].plot(kind='barh', grid=True)
```

```python
seoul['인구수'].sort_values().plot(kind='barh', grid=True) 
```

```python
seoul['외국인비율'].sort_values().plot(kind='barh', grid=True)
```

```python
seoul['고령자비율'].sort_values().plot(kind='barh', grid=True)
```

```python
top_seoul = seoul.sort_values(by='인구수', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

x = top_seoul.index # 구
y = top_seoul['인구수']

plt.figure(figsize=(10,7))
plt.title("서울시 인구수 Top 10")
plt.xlabel('구')
plt.ylabel('인구수')

plt.bar(x, y)
plt.grid()
plt.show()
plt.savefig("서울시인구현황Top10.png")
```
