```python
import matplotlib.pyplot as plt
import numpy as np
```

## 기본 그래프 plot

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그려보기
x = np.arange(0, 12, 0.01)  # 0-12까지 0.01 간격
print(x)
y = np.sin(x) # sin 함수 그리기, x 값 넣어서
```

```python
plt.plot(x,y) # plt = matplotlib, plot 그래프 그리기
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title("sin graph")
plt.show()
```

## Grid 설정

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

X, Y 라벨

```python
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Amplitude')  # 진폭
plt.show()
```

```python
plt.plot(x,y)
plt.plot(x,y-1)
plt.plot(x,y-2) # 그래프 여러 개 추가
plt.xlabel('Time')
plt.ylabel('Amplitude')  # 진폭
plt.show()
```

## 그래프 두 개 글 ㅣ기

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.grid()
# plt.legend()
plt.legend(loc=(1.05, 0.05)) # loc=Location String 줄임말로 범례의 위치 이동 가능
# plt.legend?
plt.xlabel('Time')
plt.ylabel('Amplitude')  # 진폭
plt.title("sin&cos")
plt.show()
```

```python
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')
plt.grid()

plt.ylim(-1.2,1.2)
plt.xlim(3,10)
plt.show()
```
