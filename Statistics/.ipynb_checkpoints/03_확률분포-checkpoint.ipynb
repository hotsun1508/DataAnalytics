## 이산형 균일 분포

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(low=1, high=6, size=10000)
```

```python
plt.hist(x, bins=10, color="skyblue")
plt.show()
```

## 베르누이 분포 bernoulli distribution

```python
x2 = np.random.binomial(n=1, p=1/3, size=1000)
```

```python
plt.hist(x2, bins=3, color='g')
plt.show()
```

## 이항 분포 binomial distribution

```python
x2 = np.random.binomial(n=10, p=1/3, size=1000)
```

```python
plt.hist(x2, bins=30, color='orange')
plt.show()
```

## 기하 분포

```python
x = np.random.geometric(p=1/3, size=10000)
```

```python
plt.hist(x, bins=30)
plt.show()
```

```python
max(x)  # 24라는 숫자가 있어서 그래프 범위가 0~25
```

## 정규분포

```python
# 개수가 적은 경우 -> 정규 분포가 아닌거 같음 ...
x = np.random.normal(loc=10, scale=6, size=100)
```

```python
# 유저 데이터의 분포를 아는 경우, 실험할 수 있는 임의의 데이터 생성 가능 !! 
x 
```

```python
plt.hist(x, bins=30, color='y')
plt.show()
```

```python
# 개수 늘리면 -> 정규 분포의 모양을 띰 !
x = np.random.normal(loc=60, scale=6, size=10000)
```

```python
plt.hist(x, bins=40, color='purple')
plt.show()
```

## 감마 분포 gamma distribution

```python
x = np.random.gamma(shape=2, scale=3, size=30)
```

```python
plt.hist(x, bins=30, color='g')
plt.show()
```

```python
x = np.random.gamma(shape=2, scale=3, size=3000)
```

```python
plt.hist(x, bins=30, color='g')
plt.show()
```

## 지수 분포 exponential distribution

```python
x = np.random.exponential(scale=3, size=30)
```

```python
plt.hist(x, bins=30, color='pink')
plt.show()
```

```python
x = np.random.exponential(scale=3, size=3000)
```

```python
plt.hist(x, bins=30, color='pink')
plt.show()
```

## 카이제곱 분포 chi-square distribution

```python
x = np.random.chisquare(df=3, size=1000)
```

```python
plt.hist(x, bins=30, color='gold')
plt.show()
```

## 베타 분포 beta distribution

```python
x = np.random.beta(a=2, b=2, size=300)
```

```python
plt.hist(x, bins=30)
plt.show()
```
