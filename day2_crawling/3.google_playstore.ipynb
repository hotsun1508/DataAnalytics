### Scroll down 코드
- 참고링크 : https://hello-bryan.tistory.com/194, https://stackoverflow.com/questions/20986631/how-can-i-scroll-a-web-page-using-selenium-webdriver-in-python

```
SCROLL_PAUSE_SEC = 1

# 스크롤 높이 가져옴
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 끝까지 스크롤 다운
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # 1초 대기
    time.sleep(SCROLL_PAUSE_SEC)

    # 스크롤 다운 후 스크롤 높이 다시 가져옴
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

```

# 보기 좋게 표로 저장
> open() 사용하는 방법 말고, pandas로 DataFrame에 저장해보자 !

````python
# pip install pandas
````

````python
import pandas as pd

containers = browser.find_elements(By.CLASS_NAME, 'RHo1pe')

data_list = []
# data = ""  #-> global 해줄 때 왜 "" 이거 사용 ? 빈 상태로 두면 에러뜨니까 data=1 이런식으로 아무거나 전역변수 지정해준 것임 

for container in containers:
    # 유저 닉네임
    user_name = container.find_element(By.CLASS_NAME, 'X5PpBb').text
    print(user_name)
    
    # 평점
    rating = container.find_element(By.CLASS_NAME, 'iXRFPc').get_attribute('aria-label')[10] # 평점 숫자만 가져와 index=10
    print(rating)
    
    # 댓글 작성 날짜
    date = container.find_element(By.CLASS_NAME, 'bp9Aid').text
    print(date)
    
    # 댓글 내용
    content = container.find_element(By.CLASS_NAME, 'h3YV2d').text
    print("==="*10)
    print(content)
    
    
    # 위에 전역 변수 지정해주었기 때문에 덮어 씌워짐
    data = {  
        "user_name" : user_name,
        "rating" : rating,
        "date" : date,
        "content" : content
    }
    
    data_list.append(data)  
    
print(data_list)
# print(data) 
# pd.DataFrame(data)   # 전역 변수 설정 안되어 있으면 여기서 error 뜸

df = pd.DataFrame(data_list) # DataFrame에는 리스트만 들어감 
df.to_csv('./playstore_comment.csv', encoding = 'utf-8-sig')  # csv 파일로 저장 
# df.to_csv('./playstore_comment{}.csv'.format(search_word), encoding = 'utf-8-sig')  # csv 파일로 저장 
````
